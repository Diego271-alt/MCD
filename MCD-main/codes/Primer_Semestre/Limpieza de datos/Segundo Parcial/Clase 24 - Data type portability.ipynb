```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({"x": np.arange(1, 11)})
```

# Etiquetas

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
le.fit(["paris", "paris", "tokyo", "amsterdam"])
```

```python
list(le.classes_)
```

```python
le.transform(["tokyo", "tokyo", "paris"])
```

# Estandarización Min-Max

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
data = [[-1, 2], [-0.5, 6], [0, 10], [1, 18]]
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(data)
```

```python
scaler.data_max_, scaler.data_min_
```

```python
print(scaler.transform(data))
```

```python
print(scaler.transform([[2, 2]]))
```

# Binarizar

```python
from sklearn.preprocessing import Binarizer
```

```python
X = [
    [1.0, -1.0,  2.0],
    [2.0,  0.0,  0.0],
    [0.0,  1.0, -1.0]
]
```

```python
transformer  = Binarizer().fit(X)
Y = transformer.transform(X)
```
