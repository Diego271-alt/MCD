```r
# install.packages("datos")
# install.packages("ggplot2")
library(datos)
library(ggplot2)
library(nlme)
```

## 1 Dimensi√≥n

### Histograma

```r
ggplot(diamantes, aes(quilate)) +
  geom_histogram() +
  ggtitle("Histograma para quilate")
```

```r
ggplot(diamantes, aes(quilate, color = corte)) + # Se agrega variable categ√≥rica a la est√©tica - Apilar la contirbuci√≥n de cada categor√≠a en el intervalo
  geom_histogram() +
  ggtitle("Histograma para quilate")
```

## Pol√≠gono de frequencia

```r
ggplot(diamantes, aes(quilate)) +
  geom_freqpoly() +
  ggtitle("Pol√≠gono de frecuencia")
```

```r
ggplot(diamantes, aes(quilate, color = corte)) +
  geom_freqpoly() +
  ggtitle("Pol√≠gono de frecuencia")
```

## 2 Dimensiones

### Diagrama de dispersi√≥n

```r
# Dentro de la est√©tica agregar color
ggplot(data = millas) +
  geom_point(mapping = aes(x = cilindrada, y = autopista, color = clase)) +
  ggtitle("Diagrama de dispersi√≥n") +
  labs(x = "Cilindrada", y = "Autopista")
```

```r
ggplot(millas, aes(traccion, autopista)) +
  geom_point() +
  ggtitle("Diagrama de dispersi√≥n") +
  labs(x = "Tracci√≥n", y = "Autopista")
```

```r
ggplot(Oxboys, aes(age, height, group = Subject, color = Subject)) + # Agrupar y colorear en funci√≥n de una variable
  geom_point() +
  geom_line() +
  ggtitle("Diagrama de dispersi√≥n agrupado")
```

```r
ggplot(Oxboys, aes(age, height, group = Subject, color = Subject)) +
  geom_line() +
  geom_smooth() + # Ajusta un modelo de regresi√≥n (por cada grupo en la capa)
  ggtitle("Diagrama de dispersi√≥n agrupado")
```

### Diagrama de dispersi√≥n

```r
ggplot(millas, aes(traccion, autopista)) +
  geom_boxplot() +
  ggtitle("Boxplot") +
  labs(x = "Tracci√≥n", y = "Autopista")
```

```r
ggplot(millas, aes(clase, autopista)) +
  geom_boxplot() +
  ggtitle("Boxplot") +
  labs(x = "Clase", y = "Autopista") +
  coord_flip() # Cambiar el eje
```

```r
ggplot(Oxboys, aes(Occasion, height)) +
  geom_boxplot() +
  geom_line(aes(group = Subject), color = "blue")
```

### Diagrama de viol√≠n

```r
ggplot(millas, aes(traccion, autopista)) +
  geom_violin() +
  ggtitle("Diagrama de viol√≠n") +
  labs(x = "Tracci√≥n", y = "Autopista")
```
