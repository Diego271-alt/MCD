```r
# install.packages("dplyr")
# install.packages("ggplot2")
library(dplyr)
library(ggplot2)
```

Los temas no modifican la gráfica ya elaborada, ayuda a una visualización más atractiva  

Cuatro componentes:  
- Los elementos
- Funcion de elementos
- Los temas (`theme()`)
- Temas completos

```r
# Tema completo por default
ggplot(mpg, aes(displ, hwy)) +
  geom_point(aes(color = factor(cyl)))
```

```r
# Agregando labels
ggplot(mpg, aes(displ, hwy)) +
  geom_point(aes(color = factor(cyl))) +
  labs(
      x = "Cilindrada del motor (litros)",
      y = "Millas por galón en carretera",
      title = "Kilometraje por cilindrada vs cilindros",
      subtitle = "Fuente: https://fueleconomy.gov",
      color = "Número de Cilindros"
  )
```

```r
# Agregando paleta de color
ggplot(mpg, aes(displ, hwy)) +
  geom_point(aes(color = factor(cyl))) +
  labs(
      x = "Cilindrada del motor (litros)",
      y = "Millas por galón en carretera",
      title = "Kilometraje por cilindrada vs cilindros",
      subtitle = "Fuente: https://fueleconomy.gov",
      color = "Número de Cilindros"
  ) +
  scale_color_brewer(type = "seq", palette = "Spectral")
```

```r
# Agregando paleta de color
ggplot(mpg, aes(displ, hwy)) +
  geom_point(aes(color = factor(cyl))) +
  labs(
      x = "Cilindrada del motor (litros)",
      y = "Millas por galón en carretera",
      title = "Kilometraje por cilindrada vs cilindros",
      subtitle = "Fuente: https://fueleconomy.gov",
      color = "Número de Cilindros"
  ) +
  scale_color_brewer(type = "seq", palette = "Spectral") +
  theme_bw() # Fondo blanco
```

```r
# Agregando paleta de color
ggplot(mpg, aes(displ, hwy)) +
  geom_point(aes(color = factor(cyl))) +
  labs(
      x = "Cilindrada del motor (litros)",
      y = "Millas por galón en carretera",
      title = "Kilometraje por cilindrada vs cilindros",
      subtitle = "Fuente: https://fueleconomy.gov",
      color = "Número de Cilindros"
  ) +
  scale_color_brewer(type = "seq", palette = "Spectral") +
  theme_bw() +
  theme(
      plot.title = element_text(
          face = "bold", size = 12,
          color = "red", hjust = 1 # 
      ),
      legend.background = element_rect(
          fill = "white",
          linewidth = 4,
          color = "white",
      ),
      plot.background = element_rect(
          fill = "grey90",
          color = "blue"
      ),
      panel.background = element_rect(
          color = "green",
      ),
      legend.justification = c(1, 1),
      legend.position = c(1, 1),
      panel.grid.major = element_line(
          color = "black",
          linewidth = 1.5,
          linetype = "dotted"
      )
  )
```

### Elementos personalizables de los temas  

#### 1. Elementos de la gráfica  
| Elemento         |                  |
|------------------|------------------|
| `plot.background`| `element_rect()` |
| `plot.title`     | `element_text()` |
| `plot.margin`    | `margin()`       |
