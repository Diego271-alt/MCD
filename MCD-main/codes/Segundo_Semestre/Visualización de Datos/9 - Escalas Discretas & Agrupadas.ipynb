### Scatter Plot

```r
head(mpg, 3)
```

```r
ggplot(mpg, aes(hwy, class)) +
  geom_point()
```

```r
# Igual a la gráfica anterior
ggplot(mpg, aes(hwy, class)) +
  geom_point() +
  scale_x_continuous() + # R infiere sobre el tipo de dato cual usar
  scale_y_discrete()
```

```r
ggplot(mpg, aes(hwy, class)) +
  geom_jitter(width = 0.5, height = 0.25)
```

### Boxplot

```r
ggplot(mpg, aes(manufacturer, hwy)) +
  geom_boxplot()
```

```r
ggplot(mpg, aes(manufacturer, hwy)) +
  geom_boxplot(width = 0.6) # reducir el ancho del boxplot
```

```r
ggplot(mpg, aes(manufacturer, hwy)) +
  geom_boxplot(width = 0.6) +
  guides(x = guide_axis(n.dodge = 3)) # Incluir un nivel para que los textos se lean mejor
```

```r
ggplot(mpg, aes(manufacturer, hwy)) +
  geom_boxplot(width = 0.6) +
  guides(x = guide_axis(angle = 270)) # Cambiar la distribucion
```

```r
ggplot(mpg, aes(hwy, manufacturer)) + # Invertir el eje
  geom_boxplot(width = 0.6)
```

### Histograma

```r
ggplot(mpg, aes(hwy)) +
  geom_histogram()
```

```r
ggplot(mpg, aes(hwy)) +
  geom_histogram(bins = 10)
```

```r
ggplot(mpg, aes(hwy)) +
  geom_bar() # No adecuada para variables numericas
```

```r
ggplot(mpg, aes(hwy)) +
  geom_bar() +
  scale_x_binned(n.breaks = 7) # Generar escalas agrupadas
```

### Conteo

```r
ggplot(mpg, aes(displ, hwy)) +
  geom_count()
```

```r
ggplot(mpg, aes(displ, hwy)) +
  geom_count() +
  scale_x_binned(n.breaks = 15) +
  scale_y_binned(n.breaks = 15)
```
