```r
#Para un texto tambien se puede usar annotate en vez de geom_text,
#con color y tipo de letra
ggplot(mpg, aes(displ, hwy)) + 
  geom_point() + 
  labs(
    x = "Cilindrada del motor (litros)", 
    y = "Millas por galón en carretera",
    title = "Kilometreje por cilindrada vs cilindros",
    subtitle = "Fuente: https://fueleconomy.gov",
    colour = "Número of cilindros"
  ) +
  annotate(geom="text", x=3, y=40, label="Scatter plot",color="red",fontface="italic")
```

```r
### Adicional a cadenas de texto labs permite utilizar expresiones
x <- seq(-3, 3, 0.01)
y <- x ^ 2
datos <- data.frame(x = x, y = y)
ggplot(datos, aes(x, y)) +
  geom_path() +
  labs(
      y = quote(f(x) == x ^ 2)
  )
```

```r
ggplot(datos, aes(x, y)) +
  geom_path() +
  labs(
      y = quote(f(x) == x ^ 2),
      title = "Función **Cuadrátrica**"  # Con título pero bold usando Markdown
  )
```

```r
ggplot(datos, aes(x, y)) +
  geom_path() +
  labs(
      y = quote(f(x) == x ^ 2),
      title = '*Función* **Cuadrátrica**'  # Con título pero bold usando Markdown
  ) +
  theme(
      plot.title = ggtext::element_markdown()
  )
```

```r
ggplot(datos, aes(x, y)) +
  geom_path() +
  labs(
      y = quote(f(x) == x ^ 2),
      title = '*Función* **Cuadrátrica**'  # Con título pero bold usando Markdown
  ) +
  theme(
      plot.title = ggtext::element_markdown()
  ) +
  annotate(
      "text", x = -1.5, y = 7.5, label = "Función Cuadrática"
  )
```
