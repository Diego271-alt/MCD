```r
# install.packages("ggplot2")
library(ggplot2)
```

### Escala de Tama√±o

```r
ggplot(mpg, aes(displ, hwy, size = cyl)) +
  geom_point()
```

```r
ggplot(mpg, aes(displ, hwy, size = cyl)) +
  geom_point() +
  scale_size(range = c(1, 3)) # Ajustar el tama√±o m√≠nimo y m√°ximo
```

```r
planetas <- data.frame(
    name = c(
        "Mercurio", "Venus", "Tierra",
        "Marte", "Jupiter", "Saturno",
        "Urano", "Neptuno"
    ),
    type = c(rep("Inner", 4), rep("Outer", 4)),
    position = 1:8,
    radius = c(2440, 6052, 6378, 3390, 71400, 60330, 25559, 24764),
    orbit = c(
        57900000, 108200000, 149600000,
        227900000, 778300000, 1427000000,
        2871000000, 4497100000
    )
)
```

```r
ggplot(planetas, aes(1, name, size = radius)) +
  geom_point()
```

```r
ggplot(planetas, aes(1, name, size = radius)) +
  geom_point() +
  scale_radius(limits = c(0, NA), range = c(0, 10)) # Personalizar circunferencias
```

```r
ggplot(mpg, aes(displ, manufacturer, size = hwy)) +
  geom_point(alpha = 0.2) +
  scale_size_binned() # Agrupar una varible continua
```

### Escala de Formas

```r
ggplot(mpg, aes(displ, hwy, shape = factor(cyl))) +
  geom_point()
```

```r
ggplot(mpg, aes(displ, hwy, shape = factor(cyl))) +
  geom_point() +
  scale_shape(solid = FALSE) # Eliminar el relleno
```

```r
df <- data.frame(
    shape = 1:25,
    x = (0:24) %%13,
    y = 2 - floor((0:24) / 13)
)
```

```r
ggplot(df, aes(x, y, shape = shape)) +
  geom_point(size = 4)+
  geom_text(aes(label = shape), nudge_y = 0.3) +
  theme_void() +
  scale_shape_identity() +
  ylim(0.8, 2.5)
```

```r
# Elegir manualmente las formas
ggplot(mpg, aes(displ, hwy, shape = factor(cyl))) +
  geom_point() +
  scale_shape_manual(
      values = c("4" = 25, "5" = 13, "6" = 8, "8" = 2)
  )
```

### Ancho de lineas

```r
ggplot(airquality, aes(x = factor(Month), y = Temp)) +
  geom_pointrange(stat = "summary")
```

```r
ggplot(airquality, aes(x = factor(Month), y = Temp)) +
  geom_pointrange(
      stat = "summary",
      linewidth = 2 # Tama√±o de la l√≠nea
  )
```

```r
ggplot(airquality, aes(x = Day, y = Temp, group = Month)) +
  geom_line(aes(linewidth = Month))
```

```r
ggplot(airquality, aes(x = Day, y = Temp, group = Month)) +
  geom_line(aes(linewidth = Month)) +
  scale_linewidth(range = c(0.5, 1)) # Intervalo para el relleno de las l√≠neas
```

```r
ggplot(economics_long, aes(
        x = date, y = value01,
        linetype = variable # Diferente forma de las l√≠neas
    )
  ) +
  geom_line()
```

```r
# Tipos de l√≠neas disponibles
lineas <- data.frame(value = letters[1:13])
ggplot(lineas, aes(linetype = value)) +
  geom_segment(
      mapping = aes(x=0, xend=1, y = value, yend = value)
  ) +
  theme(
      panel.grid = element_blank()
  ) +
  scale_x_continuous(NULL, NULL)
```
