```r
# install.packages("datos")
# install.packages("ggplot2")
# install.packages("hexbin")
# install.packages("dplyr")
library(datos)
library(ggplot2)
library(dplyr)
```

```r
datos <- data.frame(x = rnorm(2000), y = rnorm(2000))
```

### Gráfica de dispersión con muchos datos

```r
ggplot(datos, aes(x, y)) + 
  geom_point()
```

```r
ggplot(datos, aes(x, y)) + 
  geom_point(shape = 1) # Círculos huecos
```

```r
ggplot(datos, aes(x, y)) + 
  geom_point(shape = ".") # Círculos pixelados
```

```r
ggplot(datos, aes(x, y)) + 
  geom_point(alpha = 1/5) # Transparencia -> 0 (más transparente) (1/5 - 5 puntos se deben sobreponer para generar 1)
```
