```r
millas99 <- mpg %>% filter(year == 1999)
millas08 <- mpg %>% filter(year == 2008)
```

```r
ggplot(millas99, aes(displ, hwy)) +
  geom_point() +
  labs(x = NULL, y = NULL)
```

```r
ggplot(millas99, aes(displ, hwy)) +
  geom_point() +
  scale_x_continuous(breaks = scales::breaks_extended()) + # Es el que usa el paquete por default
  labs(x = NULL, y = NULL)
```

```r
ggplot(millas99, aes(displ, hwy)) +
  geom_point() +
  scale_x_continuous(breaks = scales::breaks_extended(n = 10)) + # 10 cortes en el eje x
  labs(x = NULL, y = NULL)
```

```r
# Cada break está asociado a una etiqueta, pero se pueden personalizar
ggplot(millas99, aes(displ, hwy)) +
  geom_point() +
  scale_x_continuous(breaks = c(2, 4, 6), labels = c("A", "B", "C")) +
  labs(x = NULL, y = NULL)
```

```r
# scales::label_bytes() # Kb, Mb, etc.
# scales::label_comma() # Decimales con comas agregadas
# scales::label_dollar() # Moneda
# scales::label_ordinal() # Con orden de clasificación
# scales::label_percent() # Porcentajes
# scales::label_pvalue() # Como valores p: < 0.05, 0.01, etc.
```

```r
ggplot(millas99, aes(displ, hwy)) +
  geom_point() +
  scale_x_continuous(labels = scales::label_dollar(prefix = "$ ", suffix = " USD")) +
  labs(x = NULL, y = NULL)
```

#### Transformación de escala

```r
ggplot(millas99, aes(displ, hwy)) +
  geom_point() +
  scale_x_reverse() # Invertir el eje x
```

```r
ggplot(millas99, aes(displ, hwy)) +
  geom_point() +
  scale_y_continuous(trans =) # Identidad
```

```r
ggplot(millas99, aes(displ, hwy)) +
  geom_point() +
  scale_x_continuous(trans = "log10") # Solo aplica el eje x, no a los datos
```

```r
ggplot(millas99, aes(log10(displ), hwy)) + # Ajustando los datos a la misma escala
  geom_point() +
  scale_x_continuous(trans = "log10")
```
