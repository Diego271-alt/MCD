```r
# install.packages("dplyr")
# install.packages("ggplot2")
# install.packages("patchwork")
library(dplyr)
library(ggplot2)
library(patchwork)
```

```r
# Histograma simple, se toma  y = __count__ (una función por defecto)
ggplot(mpg, aes(x = displ)) +
  geom_histogram()
```

```r
ggplot(mpg, aes(x = displ, y = after_stat(count))) + # Elegir una función personalizada
  geom_histogram()
```

```r
millas99 <- mpg %>% filter(year == 1999)
millas08 <- mpg %>% filter(year == 2008)
```

```r
g1 <- ggplot(millas99, aes(displ, hwy)) +
  geom_point()
g2 <- ggplot(millas08, aes(displ, hwy)) +
  geom_point()
```

```r
g1 + g2 # Concatenar gráficas
# Sin embargo, estas gráficas tienen escalas disintas, lo que puede generar cierto ruido al momento de realizar comparaciones
```

```r
g1_ajusted <- g1 +
  scale_x_continuous(limits = c(1, 7)) +
  scale_y_continuous(limits = c(10, 45))
g2_ajusted <- g2 +
  scale_x_continuous(limits = c(1, 7)) +
  scale_y_continuous(limits = c(10, 45))
```

```r
g1_ajusted + g2_ajusted
```

```r
g11 <- g1 +
  lims(x = c(1, 7), y = c(10, 45))
g22 <- g2 +
  lims(x = c(1, 7), y = c(10, 45))
```

```r
g11 + g22
```

```r
g1 +
  geom_hline(yintercept = 28, color = "red")
```

```r
g1 +
  geom_hline(yintercept = 28, color = "red") +
  coord_cartesian(ylim = c(20, 35)) # Acercamientos
```

```r
g1 +
  geom_hline(yintercept = 28, color = "red") +
  ylim(20, 35) # También produce un acercamiento, sin embargo, esto puede generar errores no deseados si se quiere tomar todo el conjunto de datos para alguna estimación
```

```r
ggplot(millas99, aes(displ, hwy)) +
  geom_point() +
  scale_x_continuous(breaks = NULL) # Elimina la cuadrícula en el eje x
```

```r
ggplot(millas99, aes(displ, hwy)) +
  geom_point() +
  scale_x_continuous(breaks = NULL) +
  labs(x = NULL, y = NULL) # Eliminar las etiquetas de los ejes
```

```r
ggplot(millas99, aes(displ, hwy)) +
  geom_point() +
  scale_x_continuous(breaks = c(2, 4.5, 6)) + # Personalizar los cortes
  scale_y_continuous(breaks = c(15, 20, 45)) +
  labs(x = NULL, y = NULL)
```
