```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/iris.csv")
```

```python
df.head(3)
```

```python
variety = df["variety"].values
```

```python
X = df.drop(["variety"], axis = 1)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 3)
```

```python
kmeans.fit(X)
```

```python
grupos = kmeans.predict(X)
```

```python
grupos
```

```python
# Tabulación cruzada
# Permite inferir que codificación es cada clase
tabulacion_cruzada = pd.crosstab(grupos, variety)
```

```python
tabulacion_cruzada
```

```python
print(kmeans.inertia_)
```

```python
print(kmeans.cluster_centers_) # Centroides
```

```python
inercias = {}
for k in range(1, 8):
    kmeans = KMeans(n_clusters = k)
    kmeans.fit(X)
    inercias[k] = kmeans.inertia_
```

```python
plt.plot(
    inercias.keys(), inercias.values(),
    label = "#Clusters vs inercia"
)
plt.legend()
plt.show()
```

De acuerdo al criterio del codo, el número de grupos óptimo es 3
