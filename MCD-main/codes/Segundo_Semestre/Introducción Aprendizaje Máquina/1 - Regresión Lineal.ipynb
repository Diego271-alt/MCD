```python
import pandas as pd
```

```python
df = pd.read_csv("data/Salary_dataset.csv")
```

```python
df.head(3)
```

```python
X = df[["YearsExperience"]].values
y = df[["Salary"]].values
```

```python
X.shape, y.shape
```

```python
# División del conjunto de datos en validación y pruebas
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 10)
```

```python
X_train.shape
```

# Regresión Lineal Simple

```python
from sklearn.linear_model import LinearRegression
regSimple = LinearRegression()
```

```python
regSimple.fit(X_train, y_train)
```

```python
predicciones = regSimple.predict(X_test)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X, y)
plt.scatter(X, regSimple.predict(X))
plt.plot(X, regSimple.predict(X))
plt.ylabel("Salario")
plt.xlabel("Años de experiencia")
plt.show()
```

## Métricas de rendimiento

```python
# R-Squared
regSimple.score(X_test, y_test)
```

```python
# RMSE
from sklearn.metrics import root_mean_squared_error
root_mean_squared_error(y_test, predicciones)
```
