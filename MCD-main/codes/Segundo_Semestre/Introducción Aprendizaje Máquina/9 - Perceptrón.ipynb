```python
import pandas as pd
```

```python
df = pd.read_csv("data/iphone_purchase_records.csv")
```

```python
gender_dummmies = pd.get_dummies(df["Gender"], drop_first = True, dtype = int)
```

```python
df = pd.concat([gender_dummmies, df.drop("Gender", axis = 1)], axis = 1)
```

```python
df.head(3)
```

```python
X = df.drop("Purchase Iphone", axis = 1)
y = df["Purchase Iphone"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size = 0.3,
    random_state = 20250522
)
```

### Perceptr√≥n

```python
from sklearn.linear_model import Perceptron
```

```python
perceptron = Perceptron(
    max_iter = 25,
    eta0 = 0.1, # Tasa de entrenamiento
    random_state = 20250522
)
```

```python
perceptron.fit(X_train, y_train)
```

```python
perceptron.score(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
y_pred = perceptron.predict(X_test)
```

```python
# Siempre predice 0
confusion_matrix(y_pred, y_test)
```

```python
print(classification_report(y_pred, y_test))
```
