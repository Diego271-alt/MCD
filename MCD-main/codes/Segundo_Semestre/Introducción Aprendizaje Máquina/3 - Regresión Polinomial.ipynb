```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/Ice_cream_selling_data.csv")
```

```python
df.head(3)
```

```python
plt.scatter(df["Temperature (°C)"], df["Ice Cream Sales (units)"])
plt.show()
```

```python
X = df[["Temperature (°C)"]].values
y = df[["Ice Cream Sales (units)"]].values
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
polinomial_model = PolynomialFeatures(degree = 2)
```

```python
X = polinomial_model.fit_transform(X)
```

```python
X
#        x^0              x                x^2
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 10)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regMult = LinearRegression()
```

```python
regMult.fit(X_train, y_train)
```

```python
predicciones = regMult.predict(X_test)
```

```python
plt.scatter(df["Temperature (°C)"], df["Ice Cream Sales (units)"])
plt.scatter(df["Temperature (°C)"], regMult.predict(X))
plt.plot(df["Temperature (°C)"], regMult.predict(X))
plt.show()
```
